Graph for the different type of anticoagulants 

```python
# import data 
df1 = pd.read_csv("anticoag.csv",thousands=',')
df1.head()
```

```python
# plotting the number of cases overtime using seaborn
plt.figure(figsize=(15,8))
ax = sns.barplot(data=df1, x="Anticoagulant", y="No. of cases", color="navy")
plt.xlabel('Type of anticoagulants reported by coroners', fontsize=15)
plt.ylabel('Number of PFDs', fontsize=15)

plt.savefig("fig2_anticoag.png", dpi=600)
```
